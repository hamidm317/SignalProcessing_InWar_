```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1024
t = np.linspace(0, 1, fs)

f = 100
x_t = np.sin(2 * f * np.pi * t) + np.random.normal(0, scale = 0.1, size = t.shape)
fb = 50
b_t = np.sin(2 * fb * np.pi * t)
```

```python
f = 100

f_s = [10, 20, 30, 40, 100, 200, 400]

x_t = np.random.normal(0, scale = 0.1 * len(f_s), size = t.shape)

for f in f_s:

    x_t = x_t + np.sin(2 * f * np.pi * t + np.random.random() * 2 * np.pi)
```

```python
fbs = np.arange(0, fs, 1)

res_s = []

for fb in fbs:

    b_t = np.sin(2 * fb * np.pi * t)
    res_m = np.sqrt(np.sum((x_t * b_t) ** 2))

    res_s.append(res_m)
```

```python
plt.plot(fbs, res_s)
```
